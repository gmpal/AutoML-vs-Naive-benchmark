```python
import os
import pandas as pd 
import numpy as np
import math
```

```python
all_models_elec_h1 = pd.read_csv("all_models_elec_h1.csv")
all_models_elec_h2 = pd.read_csv("all_models_elec_h2.csv")
all_models_exch_h1 = pd.read_csv("all_models_exch_h1.csv")
all_models_exch_h2 = pd.read_csv("all_models_exch_h2.csv")
```

```python
ds_list = [all_models_elec_h1,all_models_elec_h2,all_models_exch_h1,all_models_exch_h2]

for ds in ds_list: 
    for column in all_models_elec_h1.columns:
        ds.clip(lower = 0.000001,inplace=True)
        ds[column] = ds[column] / ds["Naive"] 
```

```python
win_loss = pd.DataFrame(columns = ds.columns)
win_loss_vec = [win_loss[:],win_loss[:],win_loss[:],win_loss[:]] 

for ds_idx in range(len(ds_list)):
    ds = ds_list[ds_idx]
    chosen = win_loss_vec[ds_idx]

    for column in ds.columns:
        counts = (ds < 1.0)[column].value_counts()
        
        chosen[column] = counts 
```

```python
win_loss = win_loss_vec[0].add(win_loss_vec[1], fill_value=0).add(win_loss_vec[2], fill_value=0).add(win_loss_vec[3], fill_value=0)
```

```python
table = win_loss.T
table.columns = ['Losses','Wins']
table = table.drop('Naive').sort_values('Wins',ascending=False)
table = table.fillna(0)
table = table.astype('int32')
table.to_csv("win_losses.csv",index=True)
```
